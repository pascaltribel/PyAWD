# Dataset generation for the Vector Acoustic Wave Equation

```python
from pyawd import VectorAcousticWaveDataset
```

## In 2 dimensions

```python
dataset_2d = VectorAcousticWaveDataset(1, nx=32, t=10, interrogators=[(-10, 0), (10, 0)], dim=2, velocity_model="Marmousi")
```

```python
dataset_2d.generate_video(0, "VAWD2D", 100)
```

```python
dataset_2d.plot_item(0)
```

```python
dataset_2d.plot_interrogators_response(0)
```

## In 3 dimensions

```python
dataset_3d = VectorAcousticWaveDataset(1, nx=32, t=10, interrogators=[(0, 10, 15), (0, -10, 15)], dim=3, velocity_model=100.)
```

```python
dataset_3d.generate_video(0, "VAWD3D", 300)
```

```python
dataset_3d.plot_interrogators_response(0)
```
